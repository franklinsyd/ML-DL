#### Compare the training times obtained by the vanilla (simple) RNN, the LSTM and the GRU models.

```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
 
# set width of bar
barWidth = 0.2

#Set plot size
plt.rcParams["figure.figsize"] = (10,6)
 
simple_rnn = [43.99, 47.88, 56.89 , 63.16 , 72.16, 97.32, 109.40, 117.77, 123.60]
gru =        [96.19, 95.76, 133.88, 126.48, 160.73, 187.12, 142.07, 184.18, 356.62]
lstm =       [99.44, 118.23, 121.87, 135.68, 153.16, 216.11, 212.70, 246.73, 395.35]
 
# Set position of bar on X axis
r1 = np.arange(len(simple_rnn))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, simple_rnn, color='r', width=barWidth, edgecolor='white', label='Simple RNN')
plt.bar(r2, lstm, color='b', width=barWidth, edgecolor='white', label='LSTM')
plt.bar(r3, gru, color='g', width=barWidth, edgecolor='white', label='GRU')

 
# Add xticks on the middle of the group bars
plt.ylabel('Training Time in Seconds', fontweight='bold')
plt.xlabel('RNN Models - No of Units (Multiclass)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(simple_rnn))], ['15', '30', '40', '60', '90', '120', '150', '180', '210'])
 
# Create legend , Save the figure & Show graphic 
fig = matplotlib.pyplot.gcf()
plt.legend()
fig.savefig('training-time-multiclass.png', dpi=300)
plt.show()

######################################### BINARY TIMES  ###############################################################

simple_rnn = [43.40, 56.94, 80.66, 59.62, 54.32, 93.61, 78.19, 78.11, 107.32, 90.63, 104.88, 105.48, 111.41, 102.25]
gru =       [60.94, 61.01, 60.53, 98.51, 120.33, 145.10, 142.81, 131.40, 161.00, 160.38, 226.48, 211.51, 276.16, 269.26 ]
lstm =  [63.41, 106.89, 162.06, 157.44, 172.34, 142.82, 137.22, 134.44, 148.54, 154.51, 238.46, 260.41, 233.76, 225.46 ]

# Set position of bar on X axis
r1 = np.arange(len(simple_rnn))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, simple_rnn, color='r', width=barWidth, edgecolor='white', label='Simple RNN')
plt.bar(r2, lstm, color='b', width=barWidth, edgecolor='white', label='LSTM')
plt.bar(r3, gru, color='g', width=barWidth, edgecolor='white', label='GRU')

 
# Add xticks on the middle of the group bars
plt.ylabel('Training Time in Seconds', fontweight='bold')
plt.xlabel('RNN Models - No of Units (Binary)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(simple_rnn))], ['15-S2', '15-R2', '30-S2', '30-R2', '40-S2', '30-R2', '60-S2',
                                                            '60-R2', '90-S2', '90-R2', '120-S2', '120-R2', '150-S2', '150-r2'])
# Create legend , Save the figure & Show graphic 
fig = matplotlib.pyplot.gcf()
plt.legend()
fig.savefig('training-time-binary.png', dpi=300)
plt.legend()
plt.show()
```
